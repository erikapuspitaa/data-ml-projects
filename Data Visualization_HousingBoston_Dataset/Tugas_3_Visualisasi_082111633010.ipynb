## **Mendeteksi Missing Value**

```python
missing_value = df.isna().sum()
print(missing_value)
```

Terdapat missing value pada kolom CRIM, ZN, INDUS, CHAS, AGE, dan LSTAT

### **Penanganan Missing Value**

```python
# Menghapus missing value
df.dropna(axis=0, inplace=True)

# Mengecek kembali missing values
missing_values_cleaned = df.isna().sum()
print("\nJumlah Missing Values setelah Pembersihan:")
print(missing_values_cleaned)
```

Sudah tidak ada missing value pada data

```python
# Menghapus kolom yang bukan numerik
df = df.drop(columns=['ZN','CHAS', 'B'])
df = df.sum()
df.head(12)
```

# **Visualisasi Data dengan Line Chart**

```python
import matplotlib.pyplot as plt

x= df.index
y= df
plt.plot(x, y)
plt.title('Line Chart')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```

### **Visualisasi Data dengan Bar Chart**

```python
df = pd.DataFrame({'Variable_Factors':df.index, 'Val':df})
ax = df.plot.bar(x='Variable_Factors', y='Val', rot=0, figsize=(12,8))
```

# **Visualisasi Data dengan Pie Chart**

```python
import matplotlib.pyplot as plt

colors = ['#B66DFF','#54B5FB','#99ff99','#ffcc99']
plt.figure(figsize = (7, 7))
plt.pie(df['Val'], labels = df['Variable_Factors'], autopct = '%.1f', colors=colors)
plt.show()
```

Dari berbagai visualisasi di atas, variable TAX memiliki nilai tertinggi dan variabel NOX memiliki nilai terendah. Artinya, nilai tarif pajak sangat mempengaruhi harga rumah dan kualitas udara lingkungan kurang mempengaruhi harga rumah di Boston

# **Visualisasi Data dengan Area Chart**

```python
data = pd.read_csv("HousingData.csv")
df =  pd.DataFrame(data)
df.isna().sum()
df = df.fillna(method='ffill')
kolom_yang_akan_dihapus = ['ZN','CHAS', 'B']
df = df.drop(kolom_yang_akan_dihapus, axis=1)
print(df.head(8))
```

```python
# Membuat diagram area
df.plot(kind='area')

# Memberi judul dan label sumbu
plt.title('Data Perumahan Boston')
plt.ylabel('Nilai Faktor yang Mempengaruhi')
plt.xlabel('Banyak data faktor')

# Menampilkan diagram
plt.show()
```

 Sumbu Y adalah nilai dari sampel data berdasarkan atribut/variabel tertentu sesuai dengan warnanya. Sedangkan sumbu X merupakan sampel data yang terdiri atas 500 record.
